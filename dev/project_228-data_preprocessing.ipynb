```python
def clean_text(text):
    if pd.isna(text):
        return ""
    text = text.lower()  # приведение к нижнему регистру
    text = re.sub(r'\d+', '', text)  # удаление цифр
    text = text.translate(str.maketrans("", "", string.punctuation))  # удаление пунктуации
    text = " ".join([word for word in text.split() if word not in russian_stopwords])  # удаление стоп-слов
    return text
```

```python
def extract_bigrams(text):
    words = text.split()
    bigrams = list(ngrams(words, 2))
    return [" ".join(bigram) for bigram in bigrams]
```

```python
import re
import pandas as pd
from natasha import Segmenter, MorphVocab, NewsEmbedding, NewsMorphTagger, Doc
from razdel import tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report

# Инициализация Natasha
segmenter = Segmenter()
morph_vocab = MorphVocab()
emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)

# Функция очистки текста
def clean_text(text):
    if pd.isna(text):
        return ""
    text = text.lower()
    text = re.sub(r'\d+', '', text)  # Удаление цифр
    text = re.sub(r'[^а-яА-Я\s]', '', text)  # Оставляем только буквы и пробелы
    text = " ".join([t.text for t in tokenize(text)])  # Токенизация razdel
    return text

# Функция для извлечения NER-сущностей
def extract_entities(text):
    doc = Doc(text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)
    doc.tag_ner()
    
    entities = {"fio": [], "date": [], "substance": [], "amount": []}
    for span in doc.spans:
        if span.type == "PER":
            entities["fio"].append(span.text)
        elif span.type == "DATE":
            entities["date"].append(span.text)
        elif "героин" in span.text.lower() or "амфетамин" in span.text.lower():
            entities["substance"].append(span.text)
        elif re.search(r'\d+\.?\d*\s*(грамм|кг|миллиграмм)', span.text):
            entities["amount"].append(span.text)
    
    return {k: ", ".join(v) for k, v in entities.items()}

# Функция для извлечения юридически значимых фраз
def extract_legal_factors(text):
    factors = {"mitigating": "", "aggravating": "", "article": "", "sentence": ""}
    
    if re.search(r'признал вину|сотрудничал со следствием', text):
        factors["mitigating"] = "да"
    if re.search(r'ранее судим|рецидив|группой лиц', text):
        factors["aggravating"] = "да"
    
    match = re.search(r'статьи?\s*(\d+\.\d+|\d+)', text)
    if match:
        factors["article"] = match.group(1)
    
    match = re.search(r'(\d+)\s*(год|месяц)', text)
    if match:
        factors["sentence"] = match.group(1) + " " + match.group(2)
    
    return factors

# Функция классификации типа преступления (ML)
def classify_crime_type(text, model, vectorizer):
    vectorized_text = vectorizer.transform([text])
    return model.predict(vectorized_text)[0]

# Основной процесс обработки датасета
def process_dataset(df):
    df["cleaned_text"] = df["result_text"].astype(str).apply(clean_text)
    
    entity_features = df["cleaned_text"].apply(extract_entities).apply(pd.Series)
    legal_features = df["cleaned_text"].apply(extract_legal_factors).apply(pd.Series)
    
    df = pd.concat([df, entity_features, legal_features], axis=1)
    
    return df

# Читаем датасет
df = pd.read_csv('cases.csv')
df = df.iloc[: , 1:]
processed_df = process_dataset(df)

# Обучение ML-модели для классификации типа преступления
X_train, X_test, y_train, y_test = train_test_split(processed_df["cleaned_text"], df["result_type"], test_size=0.2, random_state=42)
vectorizer = TfidfVectorizer(max_features=500)
classifier = RandomForestClassifier(n_estimators=100, random_state=42)

pipeline = Pipeline([
    ('vectorizer', vectorizer),
    ('classifier', classifier)
])

pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
print(classification_report(y_test, y_pred))

# Добавляем предсказанный тип преступления в датасет
processed_df["crime_type"] = processed_df["cleaned_text"].apply(lambda x: classify_crime_type(x, pipeline, vectorizer))

# Сохранение обработанного датасета
processed_df.to_csv("/mnt/data/processed_dataset.csv", index=False)
```
