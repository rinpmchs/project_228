```python
!pip install -q pandas
import re, pandas as pd

SRC = '/content/cases_random_7kto8k.csv'
DST = '/content/predicted_results_7kto8k_with_grams.csv'
df  = pd.read_csv(SRC)


GRAM_RE = re.compile(
    r'(?P<num>\d{1,4}(?:[.,]\s*\d{1,3})?)\s*'
    r'(?:грамм(?:а|ов)?|гр\.?|г\.?)\b',
    flags=re.I
)


def is_mass(s: str) -> bool:
    s = s.replace(',', '.')
    if '.' in s:
        return True
    try:
        return 0 < int(s) < 1000
    except ValueError:
        return False

DOT = re.compile(r'[.;\n\r]')


def extract(text: str):
    if pd.isna(text):
        return None
    txt = str(text)


    for m in GRAM_RE.finditer(txt):
        num_raw = m.group('num')
        if not is_mass(num_raw):
            continue

        stop = DOT.search(txt, m.end())
        scope = txt[m.start(): (stop.start() if stop else len(txt))]

        grams = []
        for g in GRAM_RE.finditer(scope):
            n = g.group('num').replace(' ', '').replace('.', ',')
            if is_mass(n) and n not in grams:
                grams.append(n)
        return ';\n'.join(grams) if grams else None

    return None


df['drug_weight'] = df['result_text'].apply(extract)
df.to_csv(DST, index=False)

display(df[['drug_weight']].head(50))
```
